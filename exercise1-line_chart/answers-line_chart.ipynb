# Imports

```python
from matplotlib import pyplot as plt
from matplotlib import rcParams
%matplotlib inline

import numpy as np
```

# Default Parameters

```python
figsize = (6, 4) # in inches
fontsize = 10
save_dpi = 300
linewidth = 0.5

rcParams.update({
    'font.size' : fontsize,
    'savefig.dpi' : 300,
    'axes.linewidth' : 0.5,
})
```

# Data

```python
# data
x = np.linspace(-2*np.pi, 2*np.pi, 50)
y1 = np.sin(x)
y2 = np.cos(x)

# labeling
xlabel = 'radians'
ylabel = 'amplitude'

# coloring
colors = ['#2b8cbe',
          '#f03b20',]
```

# Beginner
### plot this
<img src="beginner-line.png" alt="beginner" width='40%' align='left'/>

```python
# create figure and axis
fig_beg, ax = plt.subplots(1, 1, figsize=figsize)

# plot waves
ax.plot(x, y1)
ax.plot(x, y2)

# set labels
ax.set_xlabel(xlabel)
ax.set_ylabel(ylabel)
```

# Advanced
### plot this
<img src="advanced-line.png" alt="beginner" width='40%' align='left'/>

```python
# create figure and axes
fig_adv, ax = plt.subplots(1, 1, figsize=figsize)

# pre-emptively create labels
labels = ['sine', 'cosine']

# iterate and plot waves
for i, y, color, label in zip(np.arange(2), [y1, y2], colors, labels):
    # plot
    ax.plot(x, y, color=color, label=label)

    # plot points, but sample every 5th point
    xx = x[::5]
    yy = y[::5]
    ax.plot(xx, yy, 
            linestyle='none',
            marker='o',
            markersize=10,
            color=color,
            alpha=0.5)
    
# move spine around
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# y ticks are too cluttered, and change label
ax.locator_params(axis='y', nbins=3)
ax.set_ylabel('amplitude')

# change x ticks to radians (pi)
pi_range = np.arange(-2, 3)
xticklabels = ['{0:d}$\pi$'.format(i) for i in pi_range];
ax.set_xticks(pi_range * np.pi)
ax.set_xticklabels(xticklabels, y=-0.1)

# create legend from labels
ax.legend(loc='upper right')
```

# Save

```python
fig_titles = ['beginner-line.png', 'advanced-line.png']
for fig, title in zip([fig_beg, fig_adv], fig_titles):
    fig.savefig(title, dpi=600, transparent=True)
```
